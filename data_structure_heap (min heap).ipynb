```python
# tree structure with min value at the very top
# parent is always smaller than the two children
# insert at the bottom from left to right
# maintain heap structure: bubble up new element by comparing to parent; when taking out top element, swap top element 
# and newly inserted element, then swap down by comparing top to its two children, swap with the smaller one
```

```python
# implementation:
# use an array as structure is well defined from top to bot
# 
```

```python
class MinIntHeap:    
    def __init__(self):
        self.heap = []    
        
    @staticmethod
    def swap(a,b):
        c = a
        a = b
        b = c
        return a,b
    
    @staticmethod
    def find_p_index(index):
        if not index%2:
            p_index = (index-2)//2
        elif index%2:
            p_index = (index-1)//2
        return p_index
    
    def is_heap(self, a_list):
        for index in reversed(range(len(a_list))):
            if not index:
                pass
            else:            
                p_item = a_list[self.find_p_index(index)]
                if p_item > a_list[index]:
                    return False
        return True
    
    def valid_position(self, index, p_index): # in use
        item  = self.heap[index]
        p_item = self.heap[p_index]
        if p_item > item:
            return False
        else:
            return True
        
    def _shiftup(self, item_index): # in use
        if not item_index:
            return
        else:  
            p_index = self.find_p_index(item_index)

            if not self.valid_position(item_index, p_index):
                self.heap[p_index],self.heap[item_index] = self.swap(self.heap[p_index],self.heap[item_index])
                self._shiftup(p_index)
            else:
                return 
        
    def heapify(self,a_list):
        self.heap = a_list
        for index in reversed(range(len(a_list))):
            if index:
                p_index = self.find_p_index(index)
                while not self.valid_position(index,p_index):                
                    self._shiftup(index)
        return self.heap       
    
    def heappop(self,a_list):
        self.heapify(a_list)
        result = a_list[0]
        a_list.pop(0)
        self.heapify(a_list)
        return result
    
    def heappush(self, a_list, item):
        a_list.append(item)
        self.heapify(a_list)
    
    def heappushpop(self, a_list, item):
        a_list.append(item)
        self.heapify(a_list)
        result = a_list[0]
        a_list.pop(0)
        self.heapify(a_list)
        return result
    
    def heapreplace(self, a_list, item):
        self.heapify(a_list)
        result = a_list[0]
        a_list.pop(0)
        a_list.append(item)
        self.heapify(a_list)
        return result
    
    def heapview(self, a_list):
        self.heapify(a_list)
        return a_list[0]
    
    if __name__ == "__main__":
        main()     
        
```
